13. Create a list, which includes countries "San Marino" and "Tuvalu". How to find all *cities*, which has a *country* field with value found in created list of countries

```python
documents = coll.find({ "country":{ "$in": ["San Marino", "Tuvalu"]}
                       })
for doc in documents:
    pprint(doc)
```

14. Fetch all **distinct** *counties* which has a *country* field starting with "Fi"

```python
documents = coll.distinct("country", { "country":{ "$regex": "^Fi", "$options": "i"}
                           })
for doc in documents:
    pprint(doc)
```

15. Fetch all *cities* ending with "s", and sort results so, that results are printed alphabetically reversed.

```python
documents = sorted(
                    coll.distinct("city", { "city":{ "$regex": "s$", "$options": "i"}}),
                    reverse = True
                    )
for doc in documents:
    pprint(doc)
```

16. Fetch all cities in *country* Finland, ending with "nki", and sort results alphabetically

```python
documents = sorted(
                    coll.distinct( "city", { "country": "Finland", "city":{ "$regex": "nki$", "$options": "i"}}),
                    reverse = True
                    )
for doc in documents:
    pprint(doc)
```

17. Fetch first 10 cities, where *city* field has a population with a value at least 5 millions

```python
documents = coll.find({"population": {"$gt":5000000}} ).limit(10)
for doc in documents:
    print(doc["city"], int(doc["population"]))
```

19. Find all movies with imdb rating more or equal to 8.5. Sort results from best imdb rating to lower.

```python
documents = coll.find ( {"imdb.rating": {"$gte": 8.5}}).sort("imdb.rating", -1)

for doc in documents:
    print(doc["title"], doc["imdb"]["rating"])
```

20. Modify previous exercise so, that results will show movies rated with at least imdb rating 9.0 or (rating 3.0 or lower)

```python
documents = coll.find( { "$or":
                       [
                        {"imdb.rating": {"$gte": 9.0}},
                        {"imdb.rating": {"$lte": 9.0}}
                        ]
                        })  #.sort("imdb.rating", -1)

for doc in documents:
    print(doc["title"], doc["imdb"]["rating"])
```

21. Fetch all movies having exact rating 9.0 and having Tom Hanks listed in cast

```python
query = {"imdb.rating": 9.0,
                            "cast": "Tom Hanks"}
docs = coll.find(query, {"title": 1, "year": 1, "imdb.rating":1, })

for doc in docs:
    print(doc["title"])
```
